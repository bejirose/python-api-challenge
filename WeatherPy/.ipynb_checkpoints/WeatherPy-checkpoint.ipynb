```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from scipy.stats import linregress
from pprint import pprint
from datetime import date
from scipy import stats

# Import API key
from api_keys import weather_api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)

# get today's date
today = date.today().strftime('%m/%d/%Y')
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(lat_range[0], lat_range[1], size=1500)
lngs = np.random.uniform(lng_range[0], lng_range[1], size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        
        cities.append(city)

# Print the city count to confirm sufficient count
city_len = len(cities)
city_len
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
# Save config information
url = "http://api.openweathermap.org/data/2.5/weather?"
print("Beginning Data Retrieval")
print("------------------------------")
counter = 1 # counter for pritning the number of cities

#empty lists to store the values for each city
temp = [] 
city_found = []
lat = []  #empty list to store the latiture retreived for each city
long = []
maxtemp = []
humidity = []
cloudiness  = []
windspeed = []
country = []
date = []
units = "Imperial"  # units "metric" brimgs celsius data, imperial brings fahrenheit, and default is kelvin

for city in cities:
    print(f'Processing Record {counter} of Set {city_len} | {city}')
    counter += 1 # counter for pritning the number of cities
   
    # Build query URL
    query_url = url + "appid=" + weather_api_key + "&units=" + units + "&q=" + city
   
    try:
        weather_response = requests.get(query_url).json()
        #print(weather_response)
        city_found.append(weather_response['name'])
        temp.append(weather_response['main']['temp']) # Get the temperature from the response
        lat.append(weather_response['coord']['lat'])
        long.append(weather_response['coord']['lon'])
        maxtemp.append(weather_response['main']['temp_max'])
        humidity.append(weather_response['main']['humidity'])
        cloudiness.append(weather_response['clouds']['all'])
        windspeed.append(weather_response['wind']['speed'])
        country.append(weather_response['sys']['country'])
        date.append(weather_response['dt'])
    except:
        print(f'City, {city} not found. Skipping...')
    
print("------------------------------")
print("Data Retrieval Complete")    
print("------------------------------")
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
# create a data frame from cities, lat, and temp
weather_dict = {
    "City": city_found,
    "Lat": lat,
    "Lng": long,
    "Max Temp" : maxtemp,
    "Humidity" : humidity,
    "Cloudiness" : cloudiness,
    "Wind Speed" : windspeed,
    "Country" : country,
    "Date" : date
}

weather_df = pd.DataFrame(weather_dict)

#Export the city data into a .csv.
weather_df.to_csv(output_data_file, index=False)
weather_df.head()
```

## Linear Regression

```python
# Create Northern and Southern Hemisphere DataFrames
northern_df = clean_city_df.loc[clean_city_df['Lat'] >= 0]
southern_df = clean_city_df.loc[clean_city_df['Lat'] < 0]
southern_df.head()
```

```python
# OPTIONAL: Create a function to create Linear Regression plots
plot_parameters = ["Max Temp", "Humidity", "Cloudiness", "Wind Speed"]
hemispehere = [northern_df, southern_df]

x_values = []
for param in plot_parameters:
    for hs in hemispehere:
        x_values = hs["Lat"]
        y_values = hs[param]
        (slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

        # Get regression values
        regress_values = x_values * slope + intercept
        
        # Create line equation string
        line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
        
        # Build a scatter plot for each data type
        plt.scatter(x_values, y_values, marker="o")
        plt.plot(x_values,regress_values,"r-")
        plt.annotate(line_eq,(x_values.values[0] + 1,y_values.values[0] + 1),fontsize=15,color="red")
        
        # Incorporate the other graph properties
        title = param + " VsLatitude"
        plt.title(title + " Linear Regression")
        plt.xlabel("Latitude")
        plt.ylabel(param)
        plt.grid(True)

        # Save the figure
        plt.savefig("output_data/" + title + ".png")

        # Print r square value
        print(f"The r-squared is: {rvalue**2}")
        
        # Show plot
        plt.show()
```

####  Northern Hemisphere - Max Temp vs. Latitude Linear Regression

```python
x_values = northern_df["Lat"]
y_values = northern_df["Max Temp"]
(slope, intercept, rvalue, pvalue, stderr) = stats.linregress(x_values, y_values)

# Get regression values
regress_values = x_values * slope + intercept
#print(regress_values)
        
# Create line equation string
line_eq = "y = " + str(round(slope,2)) + "x +" + str(round(intercept,2))
        
# Build a scatter plot for each data type
plt.scatter(x_values, y_values, marker="o")
plt.plot(x_values,regress_values,"r-")
plt.annotate(line_eq,(x_values.values[0] + 1,y_values.values[0] + 1),fontsize=15,color="red")
        
# Incorporate the other graph properties
title = "Northern-MaxTempsVsLatitude"
plt.title(title + " Linear Regression")
plt.xlabel("Latitude")
plt.ylabel("Max Temp (F)")
plt.grid(True)

# Save the figure
plt.savefig("output_data/" + title + ".png")

# Print r square value
print(f"The r-squared is: {rvalue**2}")
        
# Show plot
plt.show()
```

# Analysis: Latitude and Max Temp in Northern Hemisphere have a weak linear relationship. The data is not within the mean and has a high variance from the mean.
